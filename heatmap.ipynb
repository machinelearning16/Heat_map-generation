```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
dataset=pd.read_csv("D:/heat_map/heat_ip.csv",header=None)
```

```python
dataset.head()
```

```python
dataset=dataset.set_index(0)
```

```python
dataset.head()
```

```python
df=dataset.T
```

```python
df.head()
```

```python
Basically we make heatmap to check correlation between features which means how much each variable
is correlated .Values which you are seeing in each block is the value of correlation between two variables
```

```python
corrMatt = df[[ 'i', 'and', 'to', 'the', 'a', 'my',
       'of', 'he', 'was', 'me', 'in','you',
        'that']].corr()
mask = np.array(corrMatt)
mask[np.tril_indices_from(mask)] = False
fig,ax= plt.subplots()
fig.set_size_inches(20,10)
sns.heatmap(corrMatt, mask=mask,vmax=.8, square=True,cmap='hot')
```

```python
fig,ax= plt.subplots()
fig.set_size_inches(20,10)
sns.heatmap(df[[ 'i', 'and', 'to', 'the', 'a', 'my',
       'of', 'he', 'was', 'me', 'in','you',
        'that']].corr(),cmap='hot',square=True)
```
